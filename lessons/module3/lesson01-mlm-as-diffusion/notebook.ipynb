# Lesson 01: Masked Language Modeling as Diffusion

**Key insight**: BERT-style masked language modeling IS a one-step diffusion process. By varying the masking rate, we get a full multi-step diffusion model.

- Masking rate = noise level
- Unmasking = denoising
- Unlike D3PM's complex transition matrices, the forward process is just "mask each token independently with probability gamma(t)"
